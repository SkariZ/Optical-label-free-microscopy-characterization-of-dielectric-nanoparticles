```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.special
import scipy.io
```

```python
mat = scipy.io.loadmat('myArray_integralSquared.mat')
data = mat['form_factor_list']  
```

```python
radius = np.linspace(1, 301, len(data))
V1=4/3*np.pi*(radius)**3
nm=1.333
n1=1.58
polarizability1= 3/2*V1*(n1**2-nm**2)/(2*nm**2+n1**2)
wavelength=532 #Wavelength in nanometers
k=2*np.pi*nm/wavelength
```

```python
print(np.abs(polarizability1).shape)
```

```python
fig, axes = plt.subplots(figsize = (8, 8))
plt.plot(radius, np.squeeze(data)*np.abs(polarizability1)**2)

axes.set_xlabel('$R$', fontsize=16)
axes.set_ylabel('$f$', fontsize=16)
#axes.set_ylim(ymin=1)
#axes.set_xlim(xmin=1)

#axes.set_xticks([0, np.pi/4, np.pi/2, 3/4 * np.pi, np.pi])
#axes.set_xticklabels(['0', '$\pi/4$', '$\pi/2$', '$3\pi/4$', '$\pi$'], fontsize = 16)
plt.show()

```

```python
fig, axes = plt.subplots(figsize = (8, 8))
plt.loglog(radius, np.squeeze(data)*np.abs(polarizability1)**2)
plt.xlabel('$R$', fontsize=16)
plt.ylabel('Integrated Signal', fontsize=16)

axes.set_ylim(ymin=0)
axes.set_xlim(xmin=0)

#fig.savefig('equation_29.svg', format = 'svg', dpi = 300, bbox_inches = 'tight', facecolor = 'white')
```

```python
k = (2 * np.pi * nm) / wavelength 
theta= np.pi
q = 2* k * np.sin(theta/2) 
Form_Factor = 3 / (q*radius)**3 *((np.sin(q*radius) - q*radius*np.cos(q*radius)))

fig, axes = plt.subplots(figsize = (8, 8))
plt.plot(radius, polarizability1*np.abs(Form_Factor))

axes.set_xlabel('$R$', fontsize=16)
axes.set_ylabel('Integrated Signal', fontsize=16)

axes.set_ylim(ymin=1)
axes.set_xlim(xmin=1)

#axes.set_xticks([0, np.pi/4, np.pi/2, 3/4 * np.pi, np.pi])
#axes.set_xticklabels(['0', '$\pi/4$', '$\pi/2$', '$3\pi/4$', '$\pi$'], fontsize = 16)
plt.show()
```

```python
fig, axes = plt.subplots(figsize = (8, 8))
plt.loglog(radius, polarizability1*np.abs(Form_Factor))
axes.set_xlabel('$R$', fontsize=16)
axes.set_ylabel('Integrated Signal', fontsize=16)
axes.set_ylim(ymin=0)
axes.set_xlim(xmin=0)
#axes.set_yticks([0, 2.8*10e4, 2.85*10e4, 2.9*10e4, 2.95*10e4])
#axes.set_yticklabels(['0', '$2.8*10^4$', '$2.85*10^4$', '$2.90*10^4$', '$2.95*10^4$'], fontsize = 16)
#fig.savefig('equation_29.svg', format = 'svg', dpi = 300, bbox_inches = 'tight', facecolor = 'white')
```

```python
radius = np.linspace(1, 301, len(data))
f = np.ones(len(data))*polarizability1
fig, axes = plt.subplots(figsize = (8, 8))
#f=np.full_like(values, polarizability1[1])
plt.plot(radius, f)
axes.set_xlabel('$R$', fontsize=16)
axes.set_ylabel('$f$', fontsize=16)

#axes.set_xticks([0, np.pi/4, np.pi/2, 3/4 * np.pi, np.pi])
#axes.set_xticklabels(['0', '$\pi/4$', '$\pi/2$', '$3\pi/4$', '$\pi$'], fontsize = 16)
plt.show()
#fig.savefig('equation_29.svg', format = 'svg', dpi = 300, bbox_inches = 'tight', facecolor = 'white')
```

```python
fig, axes = plt.subplots(figsize = (8, 8))
plt.loglog(radius, f)
axes.set_xlabel('$R$', fontsize=16)
axes.set_ylabel('$f$', fontsize=16)
axes.set_ylim(ymin=0)
axes.set_xlim(xmin=0)
#fig.savefig('equation_29.svg', format = 'svg', dpi = 300, bbox_inches = 'tight', facecolor = 'white')
```

```python
fig, axes = plt.subplots(figsize = (8, 8))

plt.semilogy(radius, np.sqrt(np.squeeze(data)*np.abs(polarizability1)**2))
plt.semilogy(radius, polarizability1*np.abs(Form_Factor))
plt.semilogy(radius, f)
axes.set_xlabel('$R[nm]$', fontsize=16)
axes.set_ylabel('Integrated Signal', fontsize=16)
axes.set_ylim(ymin=1)
axes.set_xlim(xmin=1)

plt.legend([' Darkfield', 'Interferometric scattering ','Quantitative field imaging']) 
fig.savefig('equation_30.svg', format = 'svg', dpi = 600, bbox_inches = 'tight', facecolor = 'white')
plt.show() 
```
